```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import math 
import random

dataDir = '/Users/still/Downloads/Measurements/'
runExpress = pd.read_csv(dataDir + '/express/3rd run/results.csv', sep=';', decimal= ',')
runNest = pd.read_csv(dataDir + '/nest/3rd run/results.csv', sep=';', decimal= ',')
runFastify = pd.read_csv(dataDir + '/fastify/3rd run/results.csv', sep=';', decimal= ',')
runKoa = pd.read_csv(dataDir + '/koa/3rd run/results.csv', sep=';', decimal= ',')
```

```python
# plot data to check for errors
fig, axes = plt.subplots(2, 4, figsize=(20, 20))
fig.suptitle('Lineplots')

sns.lineplot(ax=axes[0,0], data=runExpress, x='minutes', y='SPD33.CH1_Power')
sns.lineplot(ax=axes[0,1], data=runNest, x='minutes', y='SPD33.CH1_Power')
sns.lineplot(ax=axes[0,2], data=runFastify, x='minutes', y='SPD33.CH1_Power')
sns.lineplot(ax=axes[0,3], data=runKoa, x='minutes', y='SPD33.CH1_Power')

sns.lineplot(ax=axes[1,0], data=runExpress, x='minutes', y='SPD33.CH2_Power')
sns.lineplot(ax=axes[1,1], data=runNest, x='minutes', y='SPD33.CH2_Power')
sns.lineplot(ax=axes[1,2], data=runFastify, x='minutes', y='SPD33.CH2_Power')
sns.lineplot(ax=axes[1,3], data=runKoa, x='minutes', y='SPD33.CH2_Power')
```

```python
runExpress = runExpress[runExpress['SPD33.CH1_Power'] >= 2.0]
runExpress = runExpress[runExpress['SPD33.CH2_Power'] >= 2.0]

runNest = runNest[runNest['SPD33.CH1_Power'] >= 2.0]
runNest = runNest[runNest['SPD33.CH2_Power'] >= 2.0]  

runFastify = runFastify[runFastify['SPD33.CH1_Power'] >= 2.0]
runFastify = runFastify[runFastify['SPD33.CH2_Power'] >= 2.0]  

runKoa = runKoa[runKoa['SPD33.CH1_Power'] >= 2.0]
runKoa = runKoa[runKoa['SPD33.CH2_Power'] >= 2.0]

# plot data to check for errors
fig, axes = plt.subplots(2, 4, figsize=(10, 10))
fig.suptitle('Lineplots')

sns.lineplot(ax=axes[0,0], data=runExpress, x='minutes', y='SPD33.CH1_Power')
sns.lineplot(ax=axes[0,1], data=runNest, x='minutes', y='SPD33.CH1_Power')
sns.lineplot(ax=axes[0,2], data=runFastify, x='minutes', y='SPD33.CH1_Power')
sns.lineplot(ax=axes[0,3], data=runKoa, x='minutes', y='SPD33.CH1_Power')

sns.lineplot(ax=axes[1,0], data=runExpress, x='minutes', y='SPD33.CH2_Power')
sns.lineplot(ax=axes[1,1], data=runNest, x='minutes', y='SPD33.CH2_Power')
sns.lineplot(ax=axes[1,2], data=runFastify, x='minutes', y='SPD33.CH2_Power')
sns.lineplot(ax=axes[1,3], data=runKoa, x='minutes', y='SPD33.CH2_Power')
```

```python
# --------- calculate area under curve ------------- #
areaExpressServer = np.trapz(runExpress['SPD33.CH1_Power'], runExpress['minutes'])
areaExpressClient = np.trapz(runExpress['SPD33.CH2_Power'], runExpress['minutes'])

areaNestServer = np.trapz(runNest['SPD33.CH1_Power'], runNest['minutes'])
areaNestClient = np.trapz(runNest['SPD33.CH2_Power'], runNest['minutes'])

areaFastifyServer = np.trapz(runFastify['SPD33.CH1_Power'], runFastify['minutes'])
areaFastifyClient = np.trapz(runFastify['SPD33.CH2_Power'], runFastify['minutes'])

areaKoaServer = np.trapz(runKoa['SPD33.CH1_Power'], runKoa['minutes'])
areaKoaClient = np.trapz(runKoa['SPD33.CH2_Power'], runKoa['minutes'])

areaExpressServer, areaNestServer, areaFastifyServer, areaKoaServer
```

```python
# create a new dataframe with the interesting subset 
#df_sampleA = data[data[types_column_name]][typeA]

# create a new dataframe with the interesting subset 
df_sample = runExpress['SPD33.CH1_Power']

# Remove outliers from the initial data (per sample) 
# use quantile outliers when we do not have normal distributed data
def remove_outliers_quantile(data_series):
    data = data_series
    q75, q25 = np.percentile(data['SPD33.CH1_Power'],[75,25])
    intr_qr = q75 - q25

    max = q75 + (1.5 * intr_qr)
    min = q25 - (1.5 * intr_qr)

    count_total = len(data)

    data = data[(data > min) & (data < max)]
    count_wo_outliers = len(data)

    outliers_count = count_total - count_wo_outliers
    print(f'number of outliers: {outliers_count} of {count_total} observations.')
    
    # remove outliers
    return data

df = remove_outliers_quantile(df_sample)
df
```
