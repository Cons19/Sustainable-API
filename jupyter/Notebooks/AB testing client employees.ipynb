```python
import pandas as pd
import numpy as np
from numpy import loadtxt
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import math 
import random
from decimal import Decimal
```

```python
repetitions_of_random_sampling = 10000
```

```python
def ABtest(setA, setB):

    if setB[0].mean() < setA[0].mean():
        temp = setA
        setA = setB
        setB = temp

    # create a new dataframe with the interesting subset 
    data_set= pd.concat([setA, setB], ignore_index=True, sort=False)
    data_set.reset_index()

    # count the number of samples in each data set 
    count_A = len(setA)
    count_B = len(setB)

    org_mean_diff = setA.mean()-setB.mean()
    if count_A != count_B: print(f"Warning: The data original data sets do not have equally many elements: Data set A has {count_A} elements and Data set B has {count_B} elements")

    total = count_A + count_B
    fact_r = math.factorial(total)

    max_combinations = Decimal(math.factorial(total)) / Decimal((math.factorial(count_A)) * Decimal(math.factorial(total-count_A)))
    max_combinations
    if max_combinations < repetitions_of_random_sampling: 
        print(f"Warning: The number of possible combinations when picking {count_A} elements from {total} elements is {max_combinations} and you have chosen {repetitions_of_random_sampling} repetitions, thus including some repetitions.")

    list_index = list(range(total))
    #list_index

    mean_diffs = []
    for i in range(repetitions_of_random_sampling):
        samples = random.sample(list_index,total)
        sampleA_index = samples[:count_A]
        sampleB_index = samples[count_A:]

        sampleA = data_set.iloc[sampleA_index]
        sampleB = data_set.iloc[sampleB_index]

        mean_diff = sampleA.mean()-sampleB.mean()
        mean_diffs.append(mean_diff)
        #print(mean_diffs)
        print("Working..." + str(i))
    
    mean_diffs = pd.Series(mean_diffs)
    
    return mean_diffs, org_mean_diff
```

```python
dataDir = '/Users/still/Downloads/Exported'

expressClientEmployeesAllPowers = loadtxt(dataDir + '/expressClientEmployeesAllPowers.txt', unpack=False).tolist()
nestClientEmployeesAllPowers = loadtxt(dataDir + '/nestClientEmployeesAllPowers.txt', unpack=False).tolist()
fastifyClientEmployeesAllPowers = loadtxt(dataDir + '/fastifyClientEmployeesAllPowers.txt', unpack=False).tolist()
koaClientEmployeesAllPowers = loadtxt(dataDir + '/koaClientEmployeesAllPowers.txt', unpack=False).tolist()

expressClientEmployeesAllPowers = pd.DataFrame(expressClientEmployeesAllPowers)
nestClientEmployeesAllPowers = pd.DataFrame(nestClientEmployeesAllPowers)
fastifyClientEmployeesAllPowers = pd.DataFrame(fastifyClientEmployeesAllPowers)
koaClientEmployeesAllPowers = pd.DataFrame(koaClientEmployeesAllPowers)
```

```python
setA = expressClientEmployeesAllPowers
setB = nestClientEmployeesAllPowers
setC = fastifyClientEmployeesAllPowers
setD = koaClientEmployeesAllPowers

# A,B
# A,C
# A,D
# B,C
# B,D
# C,D
```

```python
res = ABtest(setA, setB)
mean_diffs = res[0]
org_mean_diff = res[1]
```

```python
mean_diffs = pd.Series(mean_diffs)
my_array = []
    
for i in range (0, len(mean_diffs)):
    my_array.append(mean_diffs[i][0])
    
counter = 0

for i in mean_diffs:
    if i[0] <= org_mean_diff[0]:
        counter += 1
        
empirical_p = counter / repetitions_of_random_sampling
    
empirical_p
#empirical_p = np.count_nonzero(pd.Series(mean_diffs) <= org_mean_diff) / repetitions_of_random_sampling
```

```python
# evaluate the histogram
fig, ax = plt.subplots(figsize=(12, 6))
values, base = np.histogram(my_array, bins=40)
plt.plot(base[:-1], values, c='gray')
# a red line indicating the original difference between the averages of the original samples 
plt.axvline(x = org_mean_diff[0], color = 'red', label = 'difference between original samples')
plt.show()
```

```python
# conclusion:
print(f"If the empirical p-value is 0 ,it means that none of the {repetitions_of_random_sampling} permuted samples "+
       f"resulted in a difference of {org_mean_diff[0]} or lower. This is only an approximation. "+
       "The probability of getting a difference in that range is not 0, but it is very small,"+
       "according to our simulation. Therefore we can reject the null hypothesis.")
```
