```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import math 
import random

dataDir = '/Users/still/Downloads/Exported'
experimentData1 = pd.read_csv(dataDir + '/expressServerArea.csv', sep=',', decimal= ',')
experimentData2 = pd.read_csv(dataDir + '/nestServerArea.csv', sep=',', decimal= ',')
experimentData3 = pd.read_csv(dataDir + '/fastifyServerArea.csv', sep=',', decimal= ',')
experimentData4 = pd.read_csv(dataDir + '/koaServerArea.csv', sep=',', decimal= ',')

experimentData1
```

```python
experimentData1['experiment']='Express'
experimentData2['experiment']='Nest'
experimentData3['experiment']='Fastify'
experimentData4['experiment']='Koa'

pdList1 = [experimentData1, experimentData2, experimentData3, experimentData4]
areasCollected = pd.concat(pdList1, ignore_index=True)
areasCollected = areasCollected.drop(columns="run")
areasCollected['index']=list(range(len(areasCollected)))
areasCollected = areasCollected[['index', 'experiment', 'area']]

pdList1
```

```python
my_pal = {"Express":"#F7CE76", "Nest":"#EEB8C5", "Fastify":"#98D4BB", "Koa":"#9AC8EB"}

areasCollected['area'] = areasCollected['area'].astype('float')

sns.set(rc={'figure.figsize':(20,5)})
sns.set_context("talk")
sns.boxplot(data=areasCollected, y='experiment' , x='area', palette=my_pal)
plt.title("Box Plot All Experiments Server")
plt.xlabel("Energy Consumption (J)")
plt.ylabel("Experiment Type")
```
