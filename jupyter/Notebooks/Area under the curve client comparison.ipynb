```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import math 
import random
```

```python
################ EMPLOYEES ################
```

```python
dataDir = '/Users/still/Downloads/Exported'

# Areas
expressClientEmployeesAreas = pd.read_csv(dataDir + '/expressClientEmployeesAreas.csv', sep=',', decimal= ',')
nestClientEmployeesAreas = pd.read_csv(dataDir + '/nestClientEmployeesAreas.csv', sep=',', decimal= ',')
fastifyClientEmployeesAreas = pd.read_csv(dataDir + '/fastifyClientEmployeesAreas.csv', sep=',', decimal= ',')
koaClientEmployeesAreas = pd.read_csv(dataDir + '/koaClientEmployeesAreas.csv', sep=',', decimal= ',')

expressClientEmployeesAreasT = pd.read_csv(dataDir + '/expressClientEmployeesAreasT.csv', sep=',', decimal= ',')
nestClientEmployeesAreasT = pd.read_csv(dataDir + '/nestClientEmployeesAreasT.csv', sep=',', decimal= ',')
fastifyClientEmployeesAreasT = pd.read_csv(dataDir + '/fastifyClientEmployeesAreasT.csv', sep=',', decimal= ',')
koaClientEmployeesAreasT = pd.read_csv(dataDir + '/koaClientEmployeesAreasT.csv', sep=',', decimal= ',')

my_pal = {"Express":"#4287f5", "Nest":"#ff9f0f", "Fastify":"#03a341", "Koa":"#fa344f"}
my_pal2 = ["#4287f5", "#ff9f0f", "#03a341", "#fa344f"]

expressClientEmployeesAreas
```

```python
for i in range (0, 31):
    expressClientEmployeesAreas['area'][i] = round(float(expressClientEmployeesAreas['area'][i]),1)
    nestClientEmployeesAreas['area'][i] = round(float(nestClientEmployeesAreas['area'][i]),1)
    fastifyClientEmployeesAreas['area'][i] = round(float(fastifyClientEmployeesAreas['area'][i]),1)
    koaClientEmployeesAreas['area'][i] = round(float(koaClientEmployeesAreas['area'][i]),1)
    expressClientEmployeesAreasT['area'][i] = round(float(expressClientEmployeesAreasT['area'][i]),1)
    nestClientEmployeesAreasT['area'][i] = round(float(nestClientEmployeesAreasT['area'][i]),1)
    fastifyClientEmployeesAreasT['area'][i] = round(float(fastifyClientEmployeesAreasT['area'][i]),1)
    koaClientEmployeesAreasT['area'][i] = round(float(koaClientEmployeesAreasT['area'][i]),1)
    
expressClientEmployeesAreas
```

```python
m1c = expressClientEmployeesAreas['area'].mean()
m2c = nestClientEmployeesAreas['area'].mean()
m3c = fastifyClientEmployeesAreas['area'].mean()
m4c = koaClientEmployeesAreas['area'].mean()
m1s = expressClientEmployeesAreasT['area'].mean()
m2s = nestClientEmployeesAreasT['area'].mean()
m3s = fastifyClientEmployeesAreasT['area'].mean()
m4s = koaClientEmployeesAreasT['area'].mean()

d = {'experiment': ["Express", "Nest", "Fastify", "Koa"],
     'meanArea': [m1c, m2c, m3c, m4c],
     'meanAreaT': [m1s, m2s, m3s, m4s]}
df = pd.DataFrame(data=d)
```

```python
df2 = df
df2 = pd.melt(df2, id_vars="experiment", var_name="means", value_name="Mean power area")
df2
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.set_context("talk")
ax = sns.barplot(x="experiment", y="Mean power area", hue="means",
            palette={'meanArea': 'white', 'meanAreaT': 'black'}, data=df2)
for bar_group, desaturate_value in zip(ax.containers, [0.5, 1]):
    for bar, color in zip(bar_group, my_pal2):
        bar.set_facecolor(sns.desaturate(color, desaturate_value))

hatches = ["///", ""]
for bars, hatch in zip(ax.containers, hatches):
    for bar in bars:
        bar.set_hatch(hatch)
        
plt.title("Energy consumption calculated in different ways for each experiment (Employees)")
plt.xlabel("Experiment Type")
plt.ylabel("Energy Consumption (J)")

handles, labels = ax.get_legend_handles_labels()

from matplotlib.legend_handler import HandlerTuple
ax.legend(handles=[tuple(bar_group) for bar_group in ax.containers],
          labels=['Trapezoid', 'Average Power * Time Interval'],
          handlelength=4, handler_map={tuple: HandlerTuple(ndivide=None, pad=0.1)})

# Add values on top of the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

plt.show()
```
