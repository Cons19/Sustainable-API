```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import math 
import random
```

```python
################ EMPLOYEES ################
```

```python
dataDir = '/Users/still/Downloads/Exported'

# Areas
expressServerEmployeesAreas = pd.read_csv(dataDir + '/expressServerEmployeesAreas.csv', sep=',', decimal= ',')
nestServerEmployeesAreas = pd.read_csv(dataDir + '/nestServerEmployeesAreas.csv', sep=',', decimal= ',')
fastifyServerEmployeesAreas = pd.read_csv(dataDir + '/fastifyServerEmployeesAreas.csv', sep=',', decimal= ',')
koaServerEmployeesAreas = pd.read_csv(dataDir + '/koaServerEmployeesAreas.csv', sep=',', decimal= ',')

expressServerEmployeesAreasT = pd.read_csv(dataDir + '/expressServerEmployeesAreasT.csv', sep=',', decimal= ',')
nestServerEmployeesAreasT = pd.read_csv(dataDir + '/nestServerEmployeesAreasT.csv', sep=',', decimal= ',')
fastifyServerEmployeesAreasT = pd.read_csv(dataDir + '/fastifyServerEmployeesAreasT.csv', sep=',', decimal= ',')
koaServerEmployeesAreasT = pd.read_csv(dataDir + '/koaServerEmployeesAreasT.csv', sep=',', decimal= ',')

my_pal = {"Express":"#4287f5", "Nest":"#ff9f0f", "Fastify":"#03a341", "Koa":"#fa344f"}
my_pal2 = ["#4287f5", "#ff9f0f", "#03a341", "#fa344f"]

expressServerEmployeesAreas
```

```python
for i in range (0, 31):
    expressServerEmployeesAreas['area'][i] = round(float(expressServerEmployeesAreas['area'][i]),1)
    nestServerEmployeesAreas['area'][i] = round(float(nestServerEmployeesAreas['area'][i]),1)
    fastifyServerEmployeesAreas['area'][i] = round(float(fastifyServerEmployeesAreas['area'][i]),1)
    koaServerEmployeesAreas['area'][i] = round(float(koaServerEmployeesAreas['area'][i]),1)
    expressServerEmployeesAreasT['area'][i] = round(float(expressServerEmployeesAreasT['area'][i]),1)
    nestServerEmployeesAreasT['area'][i] = round(float(nestServerEmployeesAreasT['area'][i]),1)
    fastifyServerEmployeesAreasT['area'][i] = round(float(fastifyServerEmployeesAreasT['area'][i]),1)
    koaServerEmployeesAreasT['area'][i] = round(float(koaServerEmployeesAreasT['area'][i]),1)
    
expressServerEmployeesAreas
```

```python
m1c = expressServerEmployeesAreas['area'].mean()
m2c = nestServerEmployeesAreas['area'].mean()
m3c = fastifyServerEmployeesAreas['area'].mean()
m4c = koaServerEmployeesAreas['area'].mean()
m1s = expressServerEmployeesAreasT['area'].mean()
m2s = nestServerEmployeesAreasT['area'].mean()
m3s = fastifyServerEmployeesAreasT['area'].mean()
m4s = koaServerEmployeesAreasT['area'].mean()

d = {'experiment': ["Express", "Nest", "Fastify", "Koa"],
     'meanArea': [m1c, m2c, m3c, m4c],
     'meanAreaT': [m1s, m2s, m3s, m4s]}
df = pd.DataFrame(data=d)
```

```python
df2 = df
df2 = pd.melt(df2, id_vars="experiment", var_name="means", value_name="Mean power area")
df2
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.set_context("talk")
ax = sns.barplot(x="experiment", y="Mean power area", hue="means",
            palette={'meanArea': 'white', 'meanAreaT': 'black'}, data=df2)
for bar_group, desaturate_value in zip(ax.containers, [0.5, 1]):
    for bar, color in zip(bar_group, my_pal2):
        bar.set_facecolor(sns.desaturate(color, desaturate_value))

hatches = ["///", ""]
for bars, hatch in zip(ax.containers, hatches):
    for bar in bars:
        bar.set_hatch(hatch)
        
plt.title("Energy consumption calculated in different ways for each experiment (Employees)")
plt.xlabel("Experiment Type")
plt.ylabel("Energy Consumption (J)")

handles, labels = ax.get_legend_handles_labels()

from matplotlib.legend_handler import HandlerTuple
ax.legend(handles=[tuple(bar_group) for bar_group in ax.containers],
          labels=['Trapezoid', 'Average Power * Time Interval'],
          handlelength=4, handler_map={tuple: HandlerTuple(ndivide=None, pad=0.1)})

# Add values on top of the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

plt.show()
```
