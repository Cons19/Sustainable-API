```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import math 
import random
```

```python
dataDir = '/Users/still/Downloads/Exported'

# Areas
expressClientEmployeesAreas = pd.read_csv(dataDir + '/expressClientEmployeesAreas.csv', sep=',', decimal= ',')
nestClientEmployeesAreas = pd.read_csv(dataDir + '/nestClientEmployeesAreas.csv', sep=',', decimal= ',')
fastifyClientEmployeesAreas = pd.read_csv(dataDir + '/fastifyClientEmployeesAreas.csv', sep=',', decimal= ',')
koaClientEmployeesAreas = pd.read_csv(dataDir + '/koaClientEmployeesAreas.csv', sep=',', decimal= ',')

koaClientEmployeesAreas
```

```python
expressClientEmployeesAreas['experiment']='Express'
nestClientEmployeesAreas['experiment']='Nest'
fastifyClientEmployeesAreas['experiment']='Fastify'
koaClientEmployeesAreas['experiment']='Koa'

areasList = [expressClientEmployeesAreas, nestClientEmployeesAreas, fastifyClientEmployeesAreas, koaClientEmployeesAreas]
areasCollectedEmployees = pd.concat(areasList, ignore_index=True)
areasCollectedEmployees = areasCollectedEmployees.drop(columns="run")
areasCollectedEmployees['index']=list(range(len(areasCollectedEmployees)))
areasCollectedEmployees = areasCollectedEmployees[['index', 'experiment', 'area']]

areasList
```

```python
color_pal = {"Express":"#4287f5", "Nest":"#ff9f0f", "Fastify":"#03a341", "Koa":"#fa344f"}

areasCollectedEmployees['area'] = areasCollectedEmployees['area'].astype('float')

sns.set(rc={'figure.figsize':(20,5)})
sns.set_context("talk")
sns.boxplot(data=areasCollectedEmployees, y='experiment' , x='area', palette=color_pal)
plt.title("All Frameworks Client Employees")
plt.xlabel("Energy Consumption (J)")
plt.ylabel("Experiment Type")
```
