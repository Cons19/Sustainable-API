```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import math 
import random
```

```python
dataDir = '/Users/still/Downloads/Exported'

# Areas
expressServerEmployeesAreas = pd.read_csv(dataDir + '/expressServerEmployeesAreas.csv', sep=',', decimal= ',')
nestServerEmployeesAreas = pd.read_csv(dataDir + '/nestServerEmployeesAreas.csv', sep=',', decimal= ',')
fastifyServerEmployeesAreas = pd.read_csv(dataDir + '/fastifyServerEmployeesAreas.csv', sep=',', decimal= ',')
koaServerEmployeesAreas = pd.read_csv(dataDir + '/koaServerEmployeesAreas.csv', sep=',', decimal= ',')

koaServerEmployeesAreas
```

```python
expressServerEmployeesAreas['experiment']='Express'
nestServerEmployeesAreas['experiment']='Nest'
fastifyServerEmployeesAreas['experiment']='Fastify'
koaServerEmployeesAreas['experiment']='Koa'

areasList = [expressServerEmployeesAreas, nestServerEmployeesAreas, fastifyServerEmployeesAreas, koaServerEmployeesAreas]
areasCollectedEmployees = pd.concat(areasList, ignore_index=True)
areasCollectedEmployees = areasCollectedEmployees.drop(columns="run")
areasCollectedEmployees['index']=list(range(len(areasCollectedEmployees)))
areasCollectedEmployees = areasCollectedEmployees[['index', 'experiment', 'area']]

areasList
```

```python
color_pal = {"Express":"#4287f5", "Nest":"#ff9f0f", "Fastify":"#03a341", "Koa":"#fa344f"}

areasCollectedEmployees['area'] = areasCollectedEmployees['area'].astype('float')

sns.set(rc={'figure.figsize':(20,5)})
sns.set_context("talk")
sns.boxplot(data=areasCollectedEmployees, y='experiment' , x='area', palette=color_pal)
plt.title("All Frameworks Server Employees")
plt.xlabel("Energy Consumption (J)")
plt.ylabel("Experiment Type")
```
